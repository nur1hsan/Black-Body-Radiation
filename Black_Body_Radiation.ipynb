```python
# Import Library yang Diperlukan

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Menuliskan Konstanta yang Digunakan

pi = math.pi
k = 1.380649 * (10**-23)      # konstanta Boltzmann dalam m^2 kg s^-2 K^-1
c = 3*(10**8)                 # kecepatan cahaya dalam m s^-1
h = 6.62607015 * (10**-34)    # konstanta Planck dalam m^2 kg s^-1
```

```python
# Mendefinisikan Fungsi Distribusi Rayleigh-Jeans, Wien, dan Planck sebagai Fungsi dari Panjang Gelombang

# Fungsi Distribusi Rayleigh-Jeans
def rj(T,l):
  return ((8*pi*k*T)/(l**4))

# Fungsi Distribusi Wien
def w(T,l):
  return (((8*pi*h*c)/(l**5))*(math.exp((-h*c)/(l*k*T))))

# Fungsi Distribusi Planck
def p(T,l):
  return ((8*pi*h*c)/((l**5)*(math.exp((h*c)/(l*k*T))-1)))
```

```python
# Inisiasi Array
Pjg_gel_6000 = []
Dist_RJ_6000 = []
Dist_W_6000 = []
Dist_P_6000 = []

Pjg_gel_10000 = []
Dist_RJ_10000 = []
Dist_W_10000 = []
Dist_P_10000 = []

Pjg_gel_30000 = []
Dist_RJ_30000 = []
Dist_W_30000 = []
Dist_P_30000 = []

T = [6000, 10000, 30000]        # Temperatur

for i in range (3):
  t = T[i]
  if (t == 6000):
    for j in range (10, 5000): # lambda dalam satuan nm
      l = j * (10**-9)          # lambda diubah ke dalam satuan meter
      n_rj = rj(t,l)
      n_w = w(t, l)
      n_p = p(t, l)
      Pjg_gel_6000.append(l)
      Dist_RJ_6000.append(n_rj)
      Dist_W_6000.append(n_w)
      Dist_P_6000.append(n_p)
  elif (t == 10000):
    for j in range (10, 5000): # lambda dalam satuan nm
      l = j * (10**-9)          # lambda diubah ke dalam satuan meter
      n_rj = rj(t,l)
      n_w = w(t, l)
      n_p = p(t, l)
      Pjg_gel_10000.append(l)
      Dist_RJ_10000.append(n_rj)
      Dist_W_10000.append(n_w)
      Dist_P_10000.append(n_p)
  else:                         # t == 30000
    for j in range (1, 5000): # lambda dalam satuan nm
      l = j * (10**-9)          # lambda diubah ke dalam satuan meter
      n_rj = rj(t,l)
      n_w = w(t, l)
      n_p = p(t, l)
      Pjg_gel_30000.append(l)
      Dist_RJ_30000.append(n_rj)
      Dist_W_30000.append(n_w)
      Dist_P_30000.append(n_p)

```

```python
# Membuat Grafik Panjang Gelombang terhadap Intensitas, pada Temperatur 6000K, 10000K, dan 30000K

plt.plot(Pjg_gel_6000, Dist_RJ_6000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_6000, Dist_W_6000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_6000, Dist_P_6000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 0.000005 ] )        
plt.ylim( [ 0, 1500000 ] )         
plt.title("6000 K")
plt.xlabel("Panjang Gelombang (m)")
plt.ylabel("Intensitas")
plt.legend(loc='upper right')
plt.show()
print()

plt.plot(Pjg_gel_10000, Dist_RJ_10000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_10000, Dist_W_10000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_10000, Dist_P_10000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 0.000002 ] )        
plt.ylim( [ 0, 20000000 ] )         
plt.title("10000 K")
plt.xlabel("Panjang Gelombang (m)")
plt.ylabel("Intensitas")
plt.legend(loc='upper right')
plt.show()
print()

plt.plot(Pjg_gel_30000, Dist_RJ_30000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_30000, Dist_W_30000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_30000, Dist_P_30000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 0.0000006 ] )        
plt.ylim( [ 0, 4500000000 ] )         
plt.title("30000 K")
plt.xlabel("Panjang Gelombang (m)")
plt.ylabel("Intensitas")
plt.legend(loc='upper right')
plt.show()
```

```python
# Mendefinisikan Fungsi Distribusi Rayleigh-Jeans, Wien, dan Planck sebagai Fungsi dari Frekuensi

# Fungsi Distribusi Rayleigh-Jeans
def rj_v(T,v):
  return ((8*pi*k*T*(v**2))/(c**3))

# Fungsi Distribusi Wien
def w_v(T,v):
  return ((8*pi*h*(v**3)*(math.exp((-h*v)/(k*T))))/(c**3))

# Fungsi Distribusi Planck
def p_v(T,v):
  return ((8*pi*h*(v**3))/((c**3)*((math.exp((h*v)/(k*T)))-1)))
```

```python
# Inisiasi Array
Pjg_gel_v_6000 = []
Dist_RJ_v_6000 = []
Dist_W_v_6000 = []
Dist_P_v_6000 = []

Pjg_gel_v_10000 = []
Dist_RJ_v_10000 = []
Dist_W_v_10000 = []
Dist_P_v_10000 = []

Pjg_gel_v_30000 = []
Dist_RJ_v_30000 = []
Dist_W_v_30000 = []
Dist_P_v_30000 = []

T = [6000, 10000, 30000]        # Temperatur

for i in range (3):
  t = T[i]
  if (t == 6000):
    for j in range (10, 10000):  # frekuensi dalam satuan Hz
      l = j * (10**12)     
      n_rj = rj_v(t,l)
      n_w = w_v(t, l)
      n_p = p_v(t, l)
      Pjg_gel_v_6000.append(l)
      Dist_RJ_v_6000.append(n_rj)
      Dist_W_v_6000.append(n_w)
      Dist_P_v_6000.append(n_p)
  elif (t == 10000):
    for j in range (10, 10000):  # frekuensi dalam satuan Hz
      l = j * (10**12)         
      n_rj = rj_v(t,l)
      n_w = w_v(t, l)
      n_p = p_v(t, l)
      Pjg_gel_v_10000.append(l)
      Dist_RJ_v_10000.append(n_rj)
      Dist_W_v_10000.append(n_w)
      Dist_P_v_10000.append(n_p)
  else:                         # t == 30000
    for j in range (10, 10000):  # frekuensi dalam satuan Hz
      l = j * (10**12)       
      n_rj = rj_v(t,l)
      n_w = w_v(t, l)
      n_p = p_v(t, l)
      Pjg_gel_v_30000.append(l)
      Dist_RJ_v_30000.append(n_rj)
      Dist_W_v_30000.append(n_w)
      Dist_P_v_30000.append(n_p)
```

```python
# Membuat Grafik Frekuensi terhadap Intensitas, pada Temperatur 6000K, 10000K, dan 30000K

plt.plot(Pjg_gel_v_6000, Dist_RJ_v_6000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_v_6000, Dist_W_v_6000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_v_6000, Dist_P_v_6000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 1750000000000000 ] )        
plt.ylim( [ 0, 0.000000000000002 ] )         
plt.title("6000 K")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Intensitas")
plt.legend(loc='upper right')
plt.show()
print()

plt.plot(Pjg_gel_v_10000, Dist_RJ_v_10000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_v_10000, Dist_W_v_10000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_v_10000, Dist_P_v_10000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 3000000000000000 ] )        
plt.ylim( [ 0, 0.00000000000002 ] )         
plt.title("10000 K")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Intensitas")
plt.legend(loc='upper right')
plt.show()
print()

plt.plot(Pjg_gel_v_30000, Dist_RJ_v_30000, c="midnightblue", label="Distribusi Rayleigh-Jeans")
plt.plot(Pjg_gel_v_30000, Dist_W_v_30000, c="purple", label="Distribusi Wien")
plt.plot(Pjg_gel_v_30000, Dist_P_v_30000, c="maroon", label="Distribusi Planck")
plt.xlim( [ 0, 7000000000000000 ] )        
plt.ylim( [ 0, 0.0000000000004 ] )         
plt.title("30000 K")
plt.xlabel("Frekuensi (Hz))")
plt.ylabel("Intensitas")
plt.legend(loc='upper left')
plt.show()
```
